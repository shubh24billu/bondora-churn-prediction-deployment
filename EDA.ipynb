```python
cols = ['BidsManual', 'Age', 'AppliedAmount',
       'Amount', 'Interest', 'LoanDuration', 'MonthlyPayment',
       'ExistingLiabilities', 'RefinanceLiabilities',
       'DebtToIncome', 'MonthlyPaymentDay',
       'PrincipalPaymentsMade', 'InterestAndPenaltyPaymentsMade',
       'PrincipalBalance', 'InterestAndPenaltyBalance',
       'NoOfPreviousLoansBeforeLoan', 'AmountOfPreviousLoansBeforeLoan',
       'PreviousRepaymentsBeforeLoan',
       'PreviousEarlyRepaymentsCountBeforeLoan', 'Target']
for col in cols:
    ax = plt.subplot()
    sns.histplot(data = df, x = col,
                multiple = 'stack', ax = ax)
    ax.set_title(f'{col}')
    plt.show()
```
